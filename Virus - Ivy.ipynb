```python
# Create a filtered dataframe from specific columns
virus_cols = ["datetime", "italy", "us"]
virus_transformed= virus_data[virus_cols].copy()

# Rename the column headers
#virus_transformed = virus_transformed.rename(columns={"License Serial Number": "id",
 #                                                         "Premises Name": "premise_name",
 #                                                         "County ID Code": "county_id"})



virus_transformed
```

```python
# Clean the data by dropping duplicates and setting the index
virus_transformed = virus_transformed.dropna()
#premise_transformed.set_index("id", inplace=True)
virus_transformed
```

In the data table, each entry indicates the infection status in the format of "#-#-#-#" -- a 4-sequel entry divided by dashes. The first sequel represents the number of confirmed cases, the second sequel represents suspected cases, the third sequel represents cured cases, the fourth sequel represents death cases.

```python
# new data frame with split value columns 
new = virus_transformed["italy"].str.split("-", n = 4, expand = True) 
  
# making separate first name column from new data frame 
virus_transformed["italy_confirmed_cases"]= new[0] 
  
# making separate last name column from new data frame 
virus_transformed["italy_suspected_cases"]= new[1] 
virus_transformed["italy_cured_cases"]= new[2] 
virus_transformed["italy_death_cases"]= new[3] 

# Dropping old Name columns 
virus_transformed.drop(columns =["italy"], inplace = True) 
  
# df display 
virus_transformed 
```

```python
new2 = virus_transformed["us"].str.split("-", n = 4, expand = True) 

virus_transformed["us_confirmed_cases"]= new2[0] 
virus_transformed["us_suspected_cases"]= new2[1] 
virus_transformed["us_cured_cases"]= new2[2] 
virus_transformed["us_death_cases"]= new2[3] 

# Dropping old Name columns 
virus_transformed.drop(columns =["us"], inplace = True) 

virus_transformed
```

```python
virus_transformed = virus_transformed.drop(virus_transformed.index[[0,88]])
virus_transformed
```
