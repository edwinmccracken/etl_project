```python
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
joined_data = pd.read_csv("joined_data.csv")
```

```python
months = mdates.MonthLocator()
months_fmt = mdates.DateFormatter('%M')
```

```python
joined_data.head()
```

```python
fig, ax = plt.subplots()
ax.plot('date', 'us_driving', data=joined_data, marker='', color='blue', linewidth=2)
ax.plot('date', 'italy_driving', data=joined_data, marker='', color='red', linewidth=2)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(months_fmt)
ax.set_xlim()
ax.grid(True)
fig.autofmt_xdate
ax.set_ylabel('Volume of Direction Requests')
ax.set_xlabel('February 1st, 2020 to April 27th, 2020')
ax.set_title('Direction Requests Driving in USA and Italy')
ax.legend()
plt.savefig('direction.png')
```
